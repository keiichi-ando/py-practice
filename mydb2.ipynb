```python
import sqlite3
import pandas as pd
from model.address import Address
```

```python
filepath = 'data/personal.txt'
_df = pd.read_table(filepath)

f_pref = lambda x: Address(x).pref
f_city = lambda x: Address(x).city

_df['都道府県'] = _df['住所'].map(f_pref)
_df['市区町村'] = _df['住所'].map(f_city)
_df.head()
```

```python
df = _df.iloc[:, [1,4,5,6]]
```

```python
db_name = 'data/sample.db'
conn = sqlite3.connect(db_name)
```

```python
df.to_sql('address', conn, if_exists="replace")
```

```python
query = "SELECT * FROM address"
c = conn.cursor()
c.execute(query)
c.fetchall()
```

```python
conn.close()
```
